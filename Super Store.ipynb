```python
import pandas as pd

# Load the dataset
data = pd.read_csv('SampleSuperstore.csv')

# Filter the data for the furniture category
furniture_data = data[data['Category'] == 'Furniture']

# Group the filtered data by sub-category and calculate the sum of sales
grouped_data = furniture_data.groupby('Sub-Category')['Sales'].sum()

# Iterate over the grouped DataFrame and print the sales for each sub-category
for sub_category, sales in grouped_data.items():
    print(f"Sub-Category: {sub_category}, Sales: {sales}")
```
